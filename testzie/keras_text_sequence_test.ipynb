```python
import keras
from keras.models import *
from keras.layers import *
from keras.optimizers import *
from keras.callbacks import *
```

```python
import numpy as np
import pandas as pd
import itertools
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.utils import to_categorical
```

```python
text = ['Being able to go from idea to result with the least possible delay is key to doing good research']
```

```python
tokenizer = Tokenizer(filters='', split=' ')
```

```python
tokenizer.fit_on_texts(text)
```

```python
vocab_size = len(tokenizer.word_index) + 1
```

```python
max_len = 10
```

```python
sequences = tokenizer.texts_to_sequences(text)
```

```python
X = []
y = []
for seq in sequences:
    for i in range(1, len(seq)):
        X.append(pad_sequences([seq[:i]], maxlen=max_len)[0])
        y.append(to_categorical(seq[i], num_classes=vocab_size))
X = np.array(X)
y = np.array(y)
```

```python
model_x = Input((10, ))
model_y = Embedding(vocab_size, 10, input_length=10)(model_x)
model_y = LSTM(32)(model_y)
model_y = Dense(vocab_size, activation='softmax')(model_y)
model = Model(model_x, model_y)
```

```python
model.compile(loss='categorical_crossentropy', optimizer='rmsprop')
```

```python
hist = model.fit(X, y, batch_size=18, epochs=500)
```

```python
plt.plot(hist.history.get('loss'))
```

```python
y_pred = np.argmax(model.predict(X), axis=1)
```
